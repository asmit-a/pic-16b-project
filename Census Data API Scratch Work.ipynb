Let's try to look at some relevant tables. First, let's consider the "Median Age At First Marriage" table. From censusreporter.org, we find that this is Table B12007. Let's try looking at this table in the default year, 2019.  

Note: I was confused about how to find the table's suffix (e.g., the \_010E in the table above), and I stumbled upon this [https://api.census.gov/data/2019/acs/acs5/groups/B12007.html]. I still don't know how exactly to find the suffixes, though. 

```python
import pandas as pd

median_age_male = pd.DataFrame(c.acs5.get(('NAME', 'B12007_001E'), {'for': 'state:*'}))
median_age_male
```

```python
# cleaning up the table
median_age_male["AGE"] = median_age_male["B12007_001E"]
median_age_male = median_age_male[["NAME", "AGE"]]
median_age_male
```

```python
median_age_male["AGE"].mean()
```

```python
median_age_female = pd.DataFrame(c.acs5.get(('NAME', 'B12007_002E'), {'for': 'state:*'}))
median_age_female
```

```python
# cleaning up the table
median_age_female["AGE"] = median_age_female["B12007_002E"]
median_age_female = median_age_female[["NAME", "AGE"]]

median_age_female["AGE"].mean()
```

```python
median_age_male_2010 = pd.DataFrame(c.acs5.get(('NAME', 'B12007_001E'), {'for': 'state:*'}, year = 2010))
median_age_male_2010
```

```python
# cleaning up the table
median_age_male_2010["AGE"] = median_age_male_2010["B12007_001E"]
median_age_male_2010 = median_age_male_2010[["NAME", "AGE"]]
# calculating mean median age across states
median_age_male_2010["AGE"].mean()
```

```python
median_age_female_2010 = pd.DataFrame(c.acs5.get(('NAME', 'B12007_002E'), {'for': 'state:*'}, year = 2010))
median_age_female_2010
```

```python
# cleaning up table
median_age_female_2010["AGE"] = median_age_female_2010["B12007_002E"]
median_age_female_2010 = median_age_female_2010[["NAME", "AGE"]]
# calculating mean median age across states
median_age_female_2010["AGE"].mean()
```
