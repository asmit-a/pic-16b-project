```python
med_age_2009 = censusdata.download('acs5', 2009,
                   censusdata.censusgeo([('state', '*')]),
                    ['B12007_001E', 'B12007_002E'])
```

```python
med_age_2009["Male"] = med_age_2009["B12007_001E"]
med_age_2009["Female"] = med_age_2009["B12007_002E"]
med_age_2009 = med_age_2009[["Male", "Female"]]
med_age_2009
```

```python
med_age_2009["Male"].mean(), med_age_2009["Female"].mean()
```
